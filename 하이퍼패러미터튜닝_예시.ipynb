## 데이터셋 불러오기

```python
import pandas as pd

df = pd.read_csv('auto.csv')
df2 = pd.get_dummies(df, columns=['origin'])
y = df2['mpg']
X = df2.drop(columns=['mpg'])
```

## 파이프라이닝

```python
from sklearn.preprocessing import StandardScaler # 데이터를 표준화
from sklearn.neighbors import KNeighborsRegressor # KNN 회귀모형
```

```python
## 파이프라인 하기 전
scaler = StandardScaler() # 표준화 객체
X_scaled = scaler.fit_transform(X) # 표준화 시키기
knn = KNeighborsRegressor(n_neighbors=5) # 하이퍼패러미터 k=5
knn.fit(X_scaled,y)
```

```python
from sklearn.pipeline import Pipeline
scaler = StandardScaler() # 표준화 객체
knn = KNeighborsRegressor(n_neighbors=5) #KNN 객체
steps = [('scaler',scaler), ('knn',knn)]
pipe = Pipeline(steps)
pipe.fit(X,y)
```

```python
pipe.predict(X)
```

## GridSearchCV를 통한 하이퍼패러미터 튜닝

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor

scaler = StandardScaler() # 데이터 표준화 객체
knn = KNeighborsRegressor() # KNN 회귀 객체
steps = [('scaler',scaler),('knn',knn)] # 파이프라인에 순서 지정
pipe = Pipeline(steps) # 파이프라인 만들기

kfold = KFold(n_splits=5, shuffle=True) # 데이터 분할 객체 생성
param_grid = {'knn__n_neighbors':range(1,20)} # 하이퍼패러미터 범위 지정, steps의 이름 참조

metrics = 'r2' # 성능측정 지표 (한개만 됨)

grid_search = GridSearchCV(estimator=pipe, # 파이프라인
                           param_grid=param_grid, # 하이퍼패러미터 범위
                           scoring=metrics, # 성능측정지표
                           cv=kfold) # 데이터 분할 객체

result = grid_search.fit(X,y)
```

```python
pd.DataFrame(result.cv_results_)
```

```python
result.best_estimator_, result.best_score_
```

## 차수변경 + 릿지회귀 하이퍼패러미터 튜닝

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

poly = PolynomialFeatures() # 데이터 다차항 추가하는 객체
reg = Ridge() # 릿지회귀 객체
steps = [('poly',poly),('reg',reg)] # 파이프라인에 순서 지정
pipe = Pipeline(steps) # 파이프라인 만들기

kfold = KFold(n_splits=5, shuffle=True) # 데이터 분할 객체 생성
param_grid = {'poly__degree':range(1,4), # 1차~3차
              'reg__alpha':[0.001, 0.01, 0.05, 0.1, 0.5, 1, 10, 100] # 람다의 범위
              } # 하이퍼패러미터 범위 지정, steps의 이름 참조

metrics = 'r2' # 성능측정 지표 (한개만 됨)

grid_search = GridSearchCV(estimator=pipe, # 파이프라인
                           param_grid=param_grid, # 하이퍼패러미터 범위
                           scoring=metrics, # 성능측정지표
                           cv=kfold) # 데이터 분할 객체

result = grid_search.fit(X,y)
```

```python
result.best_estimator_
```

```python
pd.DataFrame(result.cv_results_)
```
