## 타이타닉 생존예측

- Pclass ==> 승객 클래스
- Name ==> 이름
- Sex ==> 성별
- Age ==> 나이
- SibSp ==> 동승자 중 배우자/형제자매의 수
- Parch ==> 동승자 중 부모/가족의 수
- Ticket ==> 티켓고유번호
- Fare ==> 운임
- Cabin ==> 선실
- Embarked ==> 승선항구
- boat ==> 생존시 사용한 구명보트
- body ==> 사망후 시신 발견시 일련번호
- home.dest ==> 목적지
- 생존여부(종속변수)

```python
from sklearn.datasets import fetch_openml
X, y = fetch_openml("titanic", version=1, as_frame=True, return_X_y=True)
```

```python
X
```

```python
y
```

## 고유값 확인하기

```python
X.nunique()
```

## 열고르기

```python
columns=['pclass','sex','age','sibsp','parch','fare','embarked']
X=X[columns]
X
```

## 누락데이터 처리하기

```python
X.isnull().sum()
```

```python
X=X.dropna(axis=0)
X
```

## 범주형 데이터 바꾸기

```python
import pandas as pd
X=pd.get_dummies(X,columns=['pclass','sex','embarked'])
```

```python
X
```

```python
X=X.drop(columns='sex_male') #성별은 한개만 있어도 됨
```

```python
X
```

## y값의 인덱스 X에 맞추기

```python
X.index
```

```python
y=y[X.index]
```

```python
y
```

```python
df=X
```

```python
df['survived']=y
```

```python
df
```

## CSV 파일로 저장

```python
df.to_csv('titanic_processed.csv', index=False)
```
