## Polynomial Term 추가하기

```python
df=pd.read_csv('simple.csv')
df
```

```python
## 객체 생성
from sklearn.preprocessing import PolynomialFeatures
transformer = PolynomialFeatures(degree=2, interaction_only=True)
```

```python
## 데이터 변환
output=transformer.fit_transform(df)
output # Array형태
```

```python
## 열이름 붙이기
columns = transformer.get_feature_names(df.columns)
columns
```

```python
## 다시 데이터프레임으로
pd.DataFrame(output,columns=columns)
```

```python
## 한번에
transformer = PolynomialFeatures(degree=2, interaction_only=False)
output=transformer.fit_transform(df)
columns = transformer.get_feature_names(df.columns)
pd.DataFrame(output,columns=columns)
```
